# Gráficos Básicos
<p>Fazer gráficos com MatplotLib pode ser muito fácil caso queira algo simples, rápido e prárico.</p>
<p>Observa-se que a utilização será com base na utilização do pyplot</p>

#### Observação:
<p>Quando tive o primeiro contato com o MatPlotLib tinha um pequeno problema. Estava muito acostumado com gráficos de planilhas, e estes tinham uma interface que facilitava muito. Um exemplo era o gráfico de pontos, onde o Excel cria uma linha de tendências ao selecionar uma caixa. Isso é sacanagem, pois quem está acostumado com isso fica um pouco perdido. Por isso tenho que deixar claro que qualquer ponto, linha ou barra deve ser feita colocando a coleção de dados. Para fazer uma linha de tendência, tem que calcular os dados e plotar no gráfico.</p>

```python
# Importando a bibliotéca necessária
import matplotlib.pyplot as plt

```

## Gráfico de Linhas - Simples
<p>Iniciando por um gráfico classico. </p>
<p>Nota-se a facilidade de se criar um gráfico de linhas simples. Para análises exploratórias rápidas, onde se busca ver a informação mas não precisa de um refinamento, esta forma já resolve.</p>

`plt.plot(x, y)`

- x: É uma variável de coleção que possui os dados do eixo X (variável independente)
- y: É uma variável de coleção que possui os dados do eixo Y (variável dependente)

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

# criando o gráfico
plt.plot(meses_do_ano, faturamento_mensal)

# Publicando o gráfico. 
plt.show()
```

## Gráfico de Linhas - 2 Gráficos de Linhas Separados
<p> As vezes é necessário adicionar mais linhas no gráfico como, por exemplo, comparar do desempenho da loja visinha.</p>
<p> Para isso é só plotar os dados da segunda loja e chamar, logo em seguida, o plt.show() para publicar os gráficos separadamente.</p>


```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]

# criando o gráfico da primeira loja
plt.plot(meses_do_ano, faturamento_mensal)

# Publicando o gráfico 1. 
plt.show()

# criando o gráfico da primeira loja
plt.plot(meses_do_ano, faturamento_mensal_loja2)

# Publicando o gráfico 2. 
plt.show()
```

## Gráfico de Linhas - Gráficos de linhas Multiplos

## Melhorando o Gráfico de Linhas - Legenda

<p>Como já temos muitas linhas em um só gráfico, devemos começar a pensar em melhorar-lo para poder compreender melhor o que está sendo colocado.</p>
<p>Iniciaremos pela legenda. Para ativá-la, deve-se colocar um <b><i>label</i></b> com o nome de identificação na plostagem da linha.</p>

`plt.plot(meses_do_ano, faturamento_mensal, label='Loja Principal')`

<p>Para aparecer a legenda no gráfico, deve-se chamar o plt.legend() antes da publicação do gráfico</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]
faturamento_mensal_loja3 = [200.00, 160.6, 184.69, 305.05, 262.23, 110.90]
faturamento_mensal_loja4 = [250.00, 150.6, 174.69, 295.05, 262.23, 120.90]

# criando o gráfico da primeira loja
# Note que foi adicionado um label que guardará o nome de indentificação que ficará na legenda.
plt.plot(meses_do_ano, faturamento_mensal, label='Loja Principal')
plt.plot(meses_do_ano, faturamento_mensal_loja2, label='Loja 2')
plt.plot(meses_do_ano, faturamento_mensal_loja3, label='Loja 3')
plt.plot(meses_do_ano, faturamento_mensal_loja4, label='Loja 4')


# ###### LEGENDA
# Ativando legendas
plt.legend()
# ######


# Publicando o gráfico. 
plt.show()
```

## Melhorando o Gráfico de Linhas - Legenda
<p>Veja como a legenda melhora a interpretação do gráfico. Claro que só ela não opera milagre. Pois a densidade das informações podem dificultar na análise do gráfico. Por hora, isso já é um grande passo. </p>

## Melhorando o Gráfico de Linhas - Alterando a escala do eixo Y
<p>Para orgnaizar os dados do eixo Y, como, por exemplo, os limites do eixo e o intervalo (escala). Para isso se faz uso do plt.ylim() e plt.yticks()</p>

`plt.ylim(linite_inferior, limite_superior)`
<p>Este comando bota o limite do eixo X deixando dentro do intervalo desejado</p>

`plt.yticks(lista_de_valores)`
<p>Este comando utiliza uma lista de valores para preencher o eixo Y.</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]
faturamento_mensal_loja3 = [200.00, 160.6, 184.69, 305.05, 262.23, 110.90]
faturamento_mensal_loja4 = [250.00, 150.6, 174.69, 295.05, 262.23, 120.90]

# criando o gráfico da primeira loja
# Note que foi adicionado um label que guardará o nome de indentificação que ficará na legenda.
plt.plot(meses_do_ano, faturamento_mensal, label='Loja Principal')
plt.plot(meses_do_ano, faturamento_mensal_loja2, label='Loja 2')
plt.plot(meses_do_ano, faturamento_mensal_loja3, label='Loja 3')
plt.plot(meses_do_ano, faturamento_mensal_loja4, label='Loja 4')


# ###### LEGENDA
# Ativando legendas
plt.legend()
# ######


# ###### TÍTULO
#  Criando um titulo para o nosso gráfico
plt.title('Gráfico de Comparação de Faturamento Entre as Lojas Filiadas - 2024')
# ######


# ###### Labels dos eixos X e Y
#  Criando um label para o nosso gráfico
plt.ylabel('Meses do ano')
plt.xlabel('Faturamento em R$')
# ######


# ###### Labels dos eixos X e Y
#  Criando um label para o nosso gráfico
plt.grid()
# ######


# ###### Alterando a escala do Eixo Y
#  Definir limime do eixo Y
plt.ylim(89, 320)
# Definir os valores que aparecerão no eixo Y (escala)
lista_valores = list(range(80, 320, 10))
plt.yticks(lista_valores)
# ######


# Publicando o gráfico. 
plt.show()
```

## Melhorando o Gráfico de Linhas - Adicionando Valores nos Pontos de Intersecção
<p>A ultima mudnaça no básico é a adição dos números de referência dentro do gráfico. No nosso exemplo, ficou muito poluido o gráfico, mas para gráficos com menos informações, estes valores podem ajudar na leitura do gráfico</p>
<p>Os valores serão adicionados, um a um no gráfico com o plt.text().</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]
faturamento_mensal_loja3 = [200.00, 160.6, 184.69, 305.05, 262.23, 110.90]
faturamento_mensal_loja4 = [250.00, 150.6, 174.69, 295.05, 262.23, 120.90]

# criando o gráfico da primeira loja
# Note que foi adicionado um label que guardará o nome de indentificação que ficará na legenda.
plt.plot(meses_do_ano, faturamento_mensal, label='Loja Principal')
plt.plot(meses_do_ano, faturamento_mensal_loja2, label='Loja 2')
plt.plot(meses_do_ano, faturamento_mensal_loja3, label='Loja 3')
plt.plot(meses_do_ano, faturamento_mensal_loja4, label='Loja 4')



# ###### LEGENDA
# Ativando legendas
plt.legend()
# ######


# ###### TÍTULO
#  Criando um titulo para o nosso gráfico
plt.title('Gráfico de Comparação de Faturamento Entre as Lojas Filiadas - 2024')
# ######


# ###### Labels dos eixos X e Y
#  Criando um label para o nosso gráfico
plt.ylabel('Meses do ano')
plt.xlabel('Faturamento em R$')
# ######


# ###### Labels dos eixos X e Y
#  Criando um label para o nosso gráfico
plt.grid()
# ######


# ###### Alterando a escala do Eixo Y
#  Definir limime do eixo Y
plt.ylim(89, 320)
# Definir os valores que aparecerão no eixo Y (escala)
lista_valores = list(range(80, 320, 10))
plt.yticks(lista_valores)
# ######


# ###### Adicionando Valores dentro do Gráfico
#  Será utilizado o for para percorrer cada valor e adicionar como texto no gráfico
for i, faturamento in enumerate(faturamento_mensal):
    plt.text(meses_do_ano[i], faturamento, f'{faturamento:.2f}', ha='right', va='bottom')
    
for i, faturamento in enumerate(faturamento_mensal_loja2):
    plt.text(meses_do_ano[i], faturamento, f'{faturamento:.2f}', ha='right', va='bottom')
    
for i, faturamento in enumerate(faturamento_mensal_loja3):
    plt.text(meses_do_ano[i], faturamento, f'{faturamento:.2f}', ha='right', va='bottom')
    
for i, faturamento in enumerate(faturamento_mensal_loja4):
    plt.text(meses_do_ano[i], faturamento, f'{faturamento:.2f}', ha='right', va='bottom')
# ######



# Publicando o gráfico. 
plt.show()
```

## Melhorando o Gráfico de Linhas - Menos é Mais
