```python
# Importando a bibliotéca necessária
import matplotlib.pyplot as plt
```

## Gráfico de Linhas - Simples
<p>Iniciando por um gráfico classico. </p>
<p>Nota-se a facilidade de se criar um gráfico de linhas simples. Para análises exploratórias rápidas, onde se busca ver a informação mas não precisa de um refinamento, esta forma já resolve.</p>

`plt.plot(x, y)`

- x: É uma variável de coleção que possui os dados do eixo X (variável independente)
- y: É uma variável de coleção que possui os dados do eixo Y (variável dependente)

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

# criando o gráfico
plt.plot(meses_do_ano, faturamento_mensal)

# Publicando o gráfico. 
plt.show()
```

## Gráfico de Linhas - 2 Gráficos de Linhas Separados
<p> As vezes é necessário adicionar mais linhas no gráfico como, por exemplo, comparar do desempenho da loja visinha.</p>
<p> Para isso é só plotar os dados da segunda loja e chamar, logo em seguida, o plt.show() para publicar os gráficos separadamente.</p>


```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]

# criando o gráfico da primeira loja
plt.plot(meses_do_ano, faturamento_mensal)

# Publicando o gráfico 1. 
plt.show()

# criando o gráfico da primeira loja
plt.plot(meses_do_ano, faturamento_mensal_loja2)

# Publicando o gráfico 2. 
plt.show()
```

## Gráfico de Linhas - Gráficos de linhas Multiplos
<p> O interessante é que se quiser que as duas linhas estajam no mesmo gráfico, então será necessário fazer uma mudança drástica! A de remover o primeiro plt.shwo().</p>
<p>Ao remover, o MatPlotLib irá entender que você está trabalhando com um único gráfico, e colocará a segunda linha nele.</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = [1, 2, 3, 4, 5, 6]
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]
faturamento_mensal_loja2 = [300.00, 170.6, 194.69, 315.05, 262.23, 100.90]
faturamento_mensal_loja3 = [200.00, 160.6, 184.69, 305.05, 262.23, 110.90]
faturamento_mensal_loja4 = [250.00, 150.6, 174.69, 295.05, 262.23, 120.90]

# criando o gráfico da primeira loja
plt.plot(meses_do_ano, faturamento_mensal)
plt.plot(meses_do_ano, faturamento_mensal_loja2)
plt.plot(meses_do_ano, faturamento_mensal_loja3)
plt.plot(meses_do_ano, faturamento_mensal_loja4)

# Publicando o gráfico. 
plt.show()
```

## Gráfico de Barras - Simples
<p>O gráfico de barras é ideal para observar dados categóricos. O que é facinante neste tipo de gráficos é a facilidade dele ficar complexo.</p>
<p>Este gráfico pode ser orientado para ficar na vertical ou orizontal.</p>

`plt.bar(x, y)`

- x: São as variáveis categóricas.
- y: São os valores de cada categoria.

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

# criando o gráfico
plt.bar(meses_do_ano, faturamento_mensal)

# Publicando o gráfico Verticalmente. 
plt.show()

# Publicando o gráfico Horizontalmente. 
plt.barh(meses_do_ano, faturamento_mensal)

# Publicando o gráfico. 
plt.show()
```

## Gráfico de Pontos - Simples
<p>Este gráfico é uma boa escolha na plotagem de duas variáveis numéricas que podem ou não ter relação. Os pontos podem estar espalhados, agrupados ou alinhados formando linhas. É muito usado em gráficos de correlação.</p>


`plt.scatter(x, y)`

- x: É uma variável de coleção que possui os dados do eixo X (variável independente)
- y: É uma variável de coleção que possui os dados do eixo Y (variável dependente)
  

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível

porcentagem_bateria = [4.79, 3.95, 3.63, 5.38, 6.26, 3.69, 4.58, 6.23, 5.36, 5.6]
tempo_funcionamento = [14.72, 12.87, 12.29, 15.87, 17.63, 12.46, 14.22, 17.53, 15.82, 16.08]

# criando o gráfico
plt.scatter(porcentagem_bateria, tempo_funcionamento)

# Publicando o gráfico.
plt.show()

```

<p>Neste gráfico há sempre a dúvida. E se queisermos plotar a linha de tendência? Como se faz?</p>
<p>Esta pergunta até parece boba, mas tem que lembrar que no excel basta selecionar um campo e este já aparece e com a fórmula calculada. Mas com o Matplotlib a linha de tendência tem que ser feita manualmente. É importante que a lista da linha de tendência possua o mesmo tamanho das outras listas.</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

porcentagem_bateria = [4.79, 3.95, 3.63, 5.38, 6.26, 3.69, 4.58, 6.23, 5.36, 5.6]
tempo_funcionamento = [14.72, 12.87, 12.29, 15.87, 17.63, 12.46, 14.22, 17.53, 15.82, 16.08]

linha_tendencia = [14.63, 12.94, 12.3, 15.82, 17.59, 12.42, 14.21, 17.53, 15.78, 16.26]

# criando o gráfico de pontos
plt.scatter(porcentagem_bateria, tempo_funcionamento)

# Criando a linha de tendência
plt.plot(porcentagem_bateria, linha_tendencia, color='red')

# Publicando o gráfico.
plt.show()
```

## Gráfico de Pizza - Simples
<p>O gráfico de pizza ou setorial é ideal em casos que a observação da divisão é preferivel. Não é indicado para os casos em que comparar a proporção ou tamanho dos setores/categorias seja importante.</p>


`plt.pie(x)`

- x: São os valores de cada categoria.

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

# criando o gráfico
plt.pie(faturamento_mensal)

# Publicando o gráfico Verticalmente. 
plt.show()
```

<p>No caso em que queira colocar os nomes das categorias, podemos utilizar o atributo labls</p>

```python
# Criando as variáveis de coleção. Usaremos lista para ser o mais simples possível
meses_do_ano = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
faturamento_mensal = [100.00, 150.6, 184.69, 300.05, 257.23, 89.90]

# criando o gráfico
plt.pie(faturamento_mensal, labels=meses_do_ano)

# Publicando o gráfico Verticalmente. 
plt.show()
```
